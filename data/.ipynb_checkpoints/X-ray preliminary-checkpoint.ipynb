```python
labels_count={}
def add_labels(ls):
    for i in ls:
        if i in labels_count:
            labels_count[i]+=1
        else:
            labels_count[i]=1
```

```python
for i in range(len(df_data_entry)):
    add_labels(df_data_entry.iloc[i]["Labels"])
```

```python
def age_group(x):
    output=""
    if x<=10:
        output="0-10"
    elif x>10 and x<=20:
        output="10-20"
    elif x>20 and x<=30:
        output="20-30"
    elif x>30 and x<=40:
        output="30-40"
    elif x>40 and x<=50:
        output="40-50"
    elif x>50 and x<=60:
        output="50-60"
    elif x>60 and x<=70:
        output="60-70"
    elif x>70 and x<=80:
        output="70-80"
    elif x>80 and x<=90:
        output="80-90"
    elif x>90 and x<=100:
        output="90-100"
    return output
```

```python
df_age['age_group']=df_age["Patient Age"].apply(age_group)
```

```python
df_age.head()
```

```python
new_df_age=pd.DataFrame(df_age.groupby("age_group")["Cardiomegaly"].sum())
```

```python
for disease in labels_count:
    new_df_age[disease]=df_age.groupby("age_group")[disease].sum()
```

```python
new_df_age.head()
```

```python
new_df_age=new_df_age.drop(["No Finding"],axis=1)
new_df_age=new_df_age.reindex_axis(sorted(new_df_age.columns), axis=1)
```

```python
new_df_age
```

```python
fig=plt.figure(figsize=(10,6))
new_df_age.T.plot.barh(stacked=True)
plt.show()
```
